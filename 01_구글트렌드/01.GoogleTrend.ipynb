```python
from pytrends.request import TrendReq
import plotly.express as px

# https://joel-helloworld.tistory.com/72
# pip install pytrends
# pip install plotly

# hl = host language, tz = timezone
# 한국어, 한국시간기준
# timezone: https://forbrains.co.uk/international_tools/earth_timezones -> KST Korea Standard Time 540
pytrends = TrendReq(hl='ko', tz=540)

keyword_list = ["코로나", "코스피"] #한번에 5개 키워드 제한

pytrends.build_payload(keyword_list, cat=0, timeframe='2020-01-20 2022-12-05', geo='KR')
data = pytrends.interest_over_time()
data = data.reset_index()

figure = px.line(data, x="date", y=keyword_list, title="구글 트렌드")
figure.show()
```
