```python
# csv 파일 불러오기
import pandas as pd
df = pd.read_csv('./data/kospi_data.csv')
# 결측치 없음
df.info()
```

```python
df.head()
```

```python
x_pre = df[('2019-01-19' < df["Date"]) & (df["Date"] < '2021-01-21')]["Date"]
y_pre = df[('2019-01-19' < df["Date"]) & (df["Date"] < '2021-01-21')]["Close"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30,20))
plt.xticks(rotation=45, size=10)
plt.xlabel('Date', fontsize=1)
ax = plt.plot(x_pre,y_pre)
total_len = len(df.index)
plt.show()
```

```python
x = df[('2020-01-19' < df["Date"]) & (df["Date"] < '2021-01-21')]["Date"]
y = df[('2020-01-19' < df["Date"]) & (df["Date"] < '2021-01-21')]["Close"]
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.figure(figsize=(10,8))
plt.xticks(rotation=50, size=5)
plt.xlabel('Date', fontsize=1)
plt.plot(x,y)

plt.show()
```
